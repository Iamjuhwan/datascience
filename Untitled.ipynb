### Data Modelling

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test,y_train, y_test= train_test_split (x,y,test_size=0.3,random_state=42)
```

### Light Gradient Boosting Machine

```python
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error

# Create an instance of the LightGBM regressor
lgb = LGBMRegressor()

# Fit the model using only the target variable from y_train and y_test
lgb.fit(x_train, y_train["Item_Store_Returns"])  # Assuming "Item_Store_Returns" is the target column name
lgb_preds = lgb.predict(x_test)

# Calculate the root mean squared error
mse = mean_squared_error(y_test["Item_Store_Returns"], lgb_preds, squared=False)
print(f"Root Mean Squared Error: {mse}")
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming test_df is your test DataFrame
label_encoder = LabelEncoder()

# Encode categorical columns "Item_ID" and "Store_ID"
test["Item_ID_encoded"] = label_encoder.fit_transform(test["Item_ID"])
test["Store_ID_encoded"] = label_encoder.fit_transform(test["Store_ID"])

# Drop the original categorical columns
test.drop(["Item_ID", "Store_ID"], axis=1, inplace=True)

# Make predictions using the preprocessed test dataset
predictions = lgb.predict(test)
```
